```python
import pyarrow as pa
import pyarrow.parquet as pq
import numpy as np
# Write the DataFrame to Parquet files with different record sizes
record_sizes = [np.random.randint(5000, 10000) for i in range(100)]
offset = 0
for i, size in enumerate(record_sizes):
    # Create a PyArrow table from the current subset of the DataFrame
    table = pa.Table.from_pandas(df.iloc[offset:offset+size])

    # Write the PyArrow table to a Parquet file
    pq.write_table(table, f'/media/6TB_Volume/DataRepo/LargeSyntheticData/tmp/example_{i}.parquet', row_group_size=size)

    # Update the offset for the next subset of the DataFrame
    offset += size
```

```python
import os
from tqdm import tqdm
def df_to_parquet(df, target_dir, chunk_size=10000, **parquet_wargs):
    """Writes pandas DataFrame to parquet format with pyarrow.

    Args:
        df: DataFrame
        target_dir: local directory where parquet files are written to
        chunk_size: number of rows stored in one chunk of parquet file. Defaults to 1000000.
    """    
    for i in tqdm(range(0, len(df), chunk_size)):
        slc = df.iloc[i : i + chunk_size]
        chunk = int(i/chunk_size)
        fname = os.path.join(target_dir, f"part_{chunk:04d}.parquet")
        slc.to_parquet(fname, engine="pyarrow", **parquet_wargs)
target_dir = '/media/6TB_Volume/DataRepo/LargeSyntheticData/tmp'
df_to_parquet(df_train, target_dir)
```

```python
import dask.dataframe as dd

# Define the desired number of rows per partition
desired_rows_per_partition = 1000

# Read the Parquet files into a Dask dataframe
ddf = dd.read_parquet('/media/6TB_Volume/DataRepo/LargeSyntheticData/tmp/*.parquet')

# Define a function to check and adjust the number of rows in each partition
# def check_rows_and_repartition(partition):
#     partition = dd.from_pandas(partition, npartitions=1)
#     if len(partition) != desired_rows_per_partition:
#         return partition.repartition(divisions=list(range(0, len(partition), desired_rows_per_partition)))
#     else:
#         return partition

def check_rows_and_repartition(partition):
    num_rows = len(partition)
    if num_rows != desired_rows_per_partition:
        num_partitions = max(1, num_rows // desired_rows_per_partition)
        partition = dd.from_pandas(partition, npartitions=1)
        partition = partition.repartition(npartitions=num_partitions)
        return partition.compute()
    else:
        return partition

# Apply the function to each partition using the `map_partitions` method
ddf = ddf.map_partitions(lambda x: check_rows_and_repartition(x)).compute()
```
